```python
from yt_dlp import YoutubeDL

URL = 'https://www.youtube.com/watch?v=7zpxgyG7eGk'
opts = {
    'cookiesfrombrowser': ('chrome', ),
    'cookiefile': 'cookies'
}
with YoutubeDL(opts) as ydl:
    info = ydl.extract_info(URL, download=True)
    # ydl.download(URL)
print(info.keys())
```
